```python
import pandas as pd
dataset = pd.read_csv("housing_price_data_set.csv")
```

```python
dataset
```

```python
dataset.head()
X = dataset.iloc[:, 3:].values
y = dataset.iloc[:, 2:3].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#now we have both predicted value and the actual value with us 
#to check how much accuracy we have achieved we will import r2score
from sklearn.metrics import r2_score
print (r2_score(y_test, y_pred)*100, 2) 
```
